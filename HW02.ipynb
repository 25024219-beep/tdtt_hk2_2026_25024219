Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math

def giai_phuong_trinh_bac_2(a, b, c):
    # Nếu a = 0, phương trình trở thành phương trình bậc 1: bx + c = 0
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phương trình có vô số nghiệm."
            else:
                return "Phương trình vô nghiệm."
        else:
            x = -c / b
            return f"Phương trình trở thành bậc 1 và có một nghiệm: x = {x}"
    
    # Nếu a != 0, tính biệt thức Delta
    delta = b**2 - 4*a*c
    
    # Biện luận theo Delta
    if delta < 0:
        return "Phương trình vô nghiệm (không có nghiệm thực)."
    elif delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép: x1 = x2 = {x}"
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        return f"Phương trình có hai nghiệm phân biệt:\nx1 = {x1}\nx2 = {x2}"

# nhập liệu 
print(" GIẢI PHƯƠNG TRÌNH BẬC 2: ax^2 + bx + c = 0 ")
try:
    # Nhập hệ số từ bàn phím và ép kiểu sang số thực (float)
    a = float(input("Nhập hệ số a: "))
    b = float(input("Nhập hệ số b: "))
    c = float(input("Nhập hệ số c: "))
    
    # Gọi hàm và in kết quả
    ket_qua = giai_phuong_trinh_bac_2(a,b,c)
    print("\nKết quả:")
    print(ket_qua)
    
except ValueError:
    print("\nLỗi: Vui lòng chỉ nhập các chữ số hợp lệ!")
```

Bài 6: Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).

```python
def tim_so_lon_nhat(a, b, c, d):
    #Giả sử 'a' là số lớn nhất, gán nó vào một biến tên là 'max_val'
    max_val = a
    
    #Lần lượt đem các số còn lại so sánh với 'max_val'
    # Nếu số nào lớn hơn 'max_val' hiện tại, ta lấy số đó làm 'max_val' mới
    if b > max_val:
        max_val = b
        
    if c > max_val:
        max_val = c
        
    if d > max_val:
        max_val = d
        
    #Trả về kết quả cuối cùng
    return max_val

#Phần nhập liệu
print("TÌM GIÁ TRỊ LỚN NHẤT TRONG 4 SỐ")
try:
    # Nhập 4 số từ bàn phím
    so_a = float(input("Nhập số a: "))
    so_b = float(input("Nhập số b: "))
    so_c = float(input("Nhập số c: "))
    so_d = float(input("Nhập số d: "))
    
    # Gọi hàm và in kết quả (vẫn dùng \n để xuống dòng cho đẹp nhé!)
    ket_qua = tim_so_lon_nhat(so_a, so_b, so_c, so_d)
    print(f"\nGiá trị lớn nhất trong 4 số ({so_a}, {so_b}, {so_c}, {so_d}) là: {ket_qua}")
    
except ValueError:
    print("\nLỗi: Vui lòng chỉ nhập các chữ số hợp lệ!")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math

def kiem_tra_diem_va_duong_tron(x_i, y_i, r, x_a, y_a):
    # Tính bình phương khoảng cách từ A đến I
    # Dùng bình phương để tránh sai số khi tính căn bậc 2 (số thực)
    khoang_cach_binh_phuong = (x_a - x_i)**2 + (y_a - y_i)**2
    ban_kinh_binh_phuong = r**2
    
    # So sánh khoảng cách với bán kính
    if khoang_cach_binh_phuong < ban_kinh_binh_phuong:
        return "Điểm A nằm TRONG đường tròn."
    elif khoang_cach_binh_phuong == ban_kinh_binh_phuong:
        return "Điểm A thuộc đường tròn."
    else:
        return "Điểm A nằm ngoài đường tròn."

# Phần nhập liệu
print("KIỂM TRA ĐIỂM VÀ ĐƯỜNG TRÒN")
try:
    print("1. Nhập thông tin đường tròn:")
    tam_x = float(input("   - Tọa độ x của tâm I: "))
    tam_y = float(input("   - Tọa độ y của tâm I: "))
    ban_kinh = float(input("   - Bán kính R (R > 0): "))
    
    if ban_kinh <= 0:
        print("\nLỗi: Bán kính đường tròn phải là một số dương!")
    else:
        print("\n2. Nhập thông tin điểm A:")
        diem_x = float(input("   - Tọa độ x của điểm A: "))
        diem_y = float(input("   - Tọa độ y của điểm A: "))
        
        # Gọi hàm và in kết quả
        ket_qua = kiem_tra_diem_va_duong_tron(tam_x, tam_y, ban_kinh, diem_x, diem_y)
        print(f"\nKết quả: {ket_qua}")
        
except ValueError:
    print("\nLỗi: Vui lòng chỉ nhập các chữ số hợp lệ!")
```

Bài 11: Tính xy (Với x, y là số thực).

```python
def tinh_luy_thua(x, y):
    # Sử dụng toán tử ** để tính x mũ y
    return x ** y

# --- Phần nhập liệu và thực thi ---
print("--- TÍNH LŨY THỪA x^y ---")
try:
    # Nhập cơ số x và số mũ y từ bàn phím
    so_x = float(input("Nhập cơ số x (số thực): "))
    so_y = float(input("Nhập số mũ y (số thực): "))
    
    # Gọi hàm và in kết quả
    ket_qua = tinh_luy_thua(so_x, so_y)
    print(f"\nKết quả: {so_x} ^ {so_y} = {ket_qua}")
    
except ValueError:
    print("\nLỗi: Vui lòng chỉ nhập các chữ số hợp lệ!")
except ZeroDivisionError:
    # Bắt lỗi trường hợp x = 0 và y là số âm (ví dụ: 0^-1 tức là 1/0)
    print("\nLỗi: Không thể tính 0 mũ số âm (chia cho 0)!")
```
