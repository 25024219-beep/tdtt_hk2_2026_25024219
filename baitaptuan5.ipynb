### W5A3: Kiểm tra số nguyên tố
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True
```

### W5A4: Kiểm tra số hoàn hảo
Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_perfect_number(n: int) -> bool:
    if n <= 1:
        return False
    total = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            total += i
    return total == n
```

### W5A5: Tìm vị trí phần tử
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1.

```python
def find_first_index(arr: list, k: int) -> int:
    for i in range(len(arr)):
        if arr[i] == k:
            return i
    return -1
```

### W5A6: Tính giai thừa
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!).

```python
def factorial(n: int) -> int:
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

### W5A7: Giả lập máy tính bỏ túi
Viết hàm giả lập một máy tính bỏ túi. Hàm thực hiện tính toán kết quả của phép toán giữa số `num1` và `num2` với toán tử `operator`.
Biết các toán tử đơn giản thực hiện phép toán lần lượt là: Phép trừ (-), Phép cộng (+), Phép chia (/), Phép nhân (*). 
Kết quả làm tròn đến chữ số thập phân thứ hai.
- Đầu vào: Hai số thực và phép toán cần thực hiện (+, -, *, /).
- Đầu ra: Kết quả phép toán.

```python
def calculator(num1: float, num2: float, operator: str):
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "Loi chia cho 0"
        result = num1 / num2
    else:
        return "Phep toan khong hop le"
    return round(result, 2)
```

### W5A8: Khoảng cách Hamming
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.

```python
def hamming_distance(x: int, y: int) -> int:
    # Sử dụng phép XOR toán tử bit thuần túy để tìm các vị trí bit khác biệt
    xor_result = x ^ y
    count = 0
    while xor_result > 0:
        if xor_result & 1:
            count += 1
        xor_result >>= 1 # Dịch bit sang phải để kiểm tra bit tiếp theo
    return count
```

### W5A9: Tổng các chữ số
Viết hàm trả về tổng các chữ số của số nguyên dương n.

```python
def sum_of_digits(n: int) -> int:
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
```

### W5A10: Hai từ đẳng cấu (Isomorphic)
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về True nếu hai từ a và b là đẳng cấu, ngược lại False. Bạn có thể giả sử hai từ đầu vào có cùng độ dài.

```python
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    
    map_a_to_b = {}
    map_b_to_a = {}
    
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            map_a_to_b[char_a] = char_b
            
        if char_b in map_b_to_a:
            if map_b_to_a[char_b] != char_a:
                return False
        else:
            map_b_to_a[char_b] = char_a
            
    return True
```

### W5A12: Chia có dư
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor * q + r và 0 <= r < |divisor|.

```python
def divide_with_remainder(dividend: int, divisor: int):
    if divisor == 0:
        return None
    q = dividend // divisor
    r = dividend % divisor
    return q, r
```

### W5A14: Thực hiện phép tính
Viết hàm compute(a, b, op) thực hiện phép tính a (+ - * /) b tùy theo ký tự op.

```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return "Loi chia cho 0"
        return a / b
    else:
        return "Phep toan khong hop le"
```

### W5A15: Kiểm tra tam giác vuông
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c) -> bool:
    if a <= 0 or b <= 0 or c <= 0:
        return False
        
    # Logic tìm cạnh lớn nhất thủ công không dùng hàm bổ trợ có sẵn
    if a > b:
        if a > c:
            hypotenuse, s1, s2 = a, b, c
        else:
            hypotenuse, s1, s2 = c, a, b
    else:
        if b > c:
            hypotenuse, s1, s2 = b, a, c
        else:
            hypotenuse, s1, s2 = c, a, b
            
    return (s1 * s1) + (s2 * s2) == (hypotenuse * hypotenuse)
```
