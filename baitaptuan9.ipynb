### W9A16: Sinh chuỗi nhị phân theo thứ tự từ điển
Sử dụng thuật toán quay lui (backtracking) để xây dựng chuỗi. Do nhánh thêm "0" luôn được ưu tiên gọi trước nhánh thêm "1", kết quả sinh ra sẽ tự động tuân theo đúng thứ tự từ điển.

```python
def generate_binary_strings(n, m):
    res = []
    def backtrack(zeros, ones, current_str):
        if zeros == 0 and ones == 0:
            res.append(current_str)
            return
        if zeros > 0:
            backtrack(zeros - 1, ones, current_str + "0")
        if ones > 0:
            backtrack(zeros, ones - 1, current_str + "1")
    backtrack(n, m, "")
    for s in res:
        print(s)
```

### W9A17: Tìm dãy con tăng nghiêm ngặt dài nhất và truy vết
Sử dụng phương pháp quy hoạch động kết hợp với mảng lưu chỉ số phần tử đứng trước để khôi phục lại toàn bộ chuỗi kết quả.

```python
def longest_strictly_increasing_subsequence(arr):
    n = len(arr)
    if n == 0:
        return []
    dp = [1] * n
    parent = [-1] * n
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j]:
                if dp[j] + 1 > dp[i]:
                    dp[i] = dp[j] + 1
                    parent[i] = j
    max_len = 0
    best_idx = 0
    for i in range(n):
        if dp[i] > max_len:
            max_len = dp[i]
            best_idx = i
    path = []
    curr = best_idx
    while curr != -1:
        path.append(arr[curr])
        curr = parent[curr]
    rev_path = []
    for i in range(len(path) - 1, -1, -1):
        rev_path.append(path[i])
    return rev_path
```

### W9A20: Chuẩn hóa chuỗi ký tự thủ công
Tìm vị trí biên để loại bỏ khoảng trắng đầu cuối, tự tách từ và chuyển đổi chữ hoa/chữ thường thông qua mã ASCII bảng chữ cái.

```python
def normalize_string(s):
    n = len(s)
    start = 0
    while start < n and s[start] == " ":
        start += 1
    end = n - 1
    while end >= start and s[end] == " ":
        end -= 1
    if start > end:
        return ""
    
    words = []
    curr_word = ""
    for i in range(start, end + 1):
        if s[i] != " ":
            curr_word += s[i]
        else:
            if curr_word != "":
                words.append(curr_word)
                curr_word = ""
    if curr_word != "":
        words.append(curr_word)
        
    res = ""
    for idx in range(len(words)):
        w = words[idx]
        capitalized = ""
        first = w[0]
        if "a" <= first <= "z":
            first = chr(ord(first) - 32)
        capitalized += first
        for i in range(1, len(w)):
            char = w[i]
            if "A" <= char <= "Z":
                char = chr(ord(char) + 32)
            capitalized += char
        res += capitalized
        if idx < len(words) - 1:
            res += " "
    return res
```

### W9A21: Thuật toán QuickSelect phân hoạch giảm dần
Tìm kiếm phần tử lớn thứ k với độ phức tạp trung bình O(n) mà không cần thực hiện sắp xếp toàn bộ danh sách.

```python
def find_kth_largest(arr, k):
    def partition(low, high):
        pivot = arr[high]
        i = low
        for j in range(low, high):
            if arr[j] >= pivot:
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
        arr[i], arr[high] = arr[high], arr[i]
        return i

    def quickselect(low, high, target_idx):
        if low == high:
            return arr[low]
        p = partition(low, high)
        if p == target_idx:
            return arr[p]
        elif p < target_idx:
            return quickselect(p + 1, high, target_idx)
        else:
            return quickselect(low, p - 1, target_idx)

    return quickselect(0, len(arr) - 1, k - 1)
```

### W9A22: So sánh tần suất ký tự
Đếm số lần xuất hiện của từng ký tự trong cả hai chuỗi và đối chiếu trực tiếp để xác định xem chúng có đồng dạng cấu trúc ký tự hay không.

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False
    counts1 = {}
    counts2 = {}
    for c in s1:
        if c in counts1:
            counts1[c] += 1
        else:
            counts1[c] = 1
    for c in s2:
        if c in counts2:
            counts2[c] += 1
        else:
            counts2[c] = 1
    for c in counts1:
        if c not in counts2 or counts1[c] != counts2[c]:
            return False
    return True
```

### W9A23: Thiết lập ngăn xếp đơn điệu giảm dần
Loại bỏ các chữ số nhỏ hơn đứng trước khi còn lượt xóa nhằm tạo ra chuỗi chữ số lớn nhất có thể từ trái qua phải mà không đổi thứ tự.

```python
def largest_number_deletions(s, k):
    n = len(s)
    to_drop = n - k
    stack = []
    for digit in s:
        while to_drop > 0 and stack and stack[len(stack) - 1] < digit:
            stack.pop()
            to_drop -= 1
        stack.append(digit)
    res = ""
    for i in range(k):
        res += stack[i]
    return res
```

### W9A2: Tìm kiếm nhị phân trên không gian giá trị ma trận

```python
def find_kth_largest_matrix(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    low = matrix[0][0]
    high = matrix[n - 1][m - 1]
    target_count = n * m - k + 1
    ans = low
    while low <= high:
        mid = (low + high) // 2
        count = 0
        r = n - 1
        c = 0
        while r >= 0 and c < m:
            if matrix[r][c] <= mid:
                count += (r + 1)
                c += 1
            else:
                r -= 1
        if count >= target_count:
            ans = mid
            high = mid - 1
        else:
            low = mid + 1
    return ans
```

### W9A3: Khảo sát hoán vị cặp chữ số tạo số nhỏ nhất

```python
def swap_two_digits_smallest_v2(num):
    s = str(num)
    n = len(s)
    min_val = num
    for i in range(n):
        for j in range(i + 1, n):
            chars = list(s)
            chars[i], chars[j] = chars[j], chars[i]
            if chars[0] != "0":
                val = 0
                for c in chars:
                    val = val * 10 + (ord(c) - ord("0"))
                if val < min_val:
                    min_val = val
    return min_val
```

### W9A4: Khớp chuỗi bằng Quy hoạch động (Dynamic Programming)
Xác định khả năng biến đổi từ xâu a sang xâu b bằng cách đưa ra quyết định giữ, xóa hoặc viết hoa các ký tự tương ứng tại mỗi trạng thái.

```python
def can_transform(a, b):
    len_a = len(a)
    len_b = len(b)
    dp = []
    for i in range(len_a + 1):
        row = [False] * (len_b + 1)
        dp.append(row)
    dp[0][0] = True
    
    for i in range(1, len_a + 1):
        char_a = a[i - 1]
        is_upper = ("A" <= char_a <= "Z")
        for j in range(len_b + 1):
            if is_upper:
                if j > 0 and char_a == b[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]
            else:
                if dp[i - 1][j]:
                    dp[i][j] = True
                elif j > 0:
                    upper_a = chr(ord(char_a) - 32)
                    if upper_a == b[j - 1] and dp[i - 1][j - 1]:
                        dp[i][j] = True
                        
    if dp[len_a][len_b]:
        return "YES"
    return "NO"
```

```python
def is_sublist(list1, list2):
    n1 = len(list1)
    n2 = len(list2)
    if n1 == 0:
        return True
    if n1 > n2:
        return False
    for i in range(n2 - n1 + 1):
        match = True
        for j in range(n1):
            if list2[i + j] != list1[j]:
                match = False
                break
        if match:
            return True
    return False
```

### W9A6: Thuật toán tìm hoán vị kế tiếp (Next Permutation)
Tìm điểm gãy giảm dần từ phải sang, hoán đổi với phần tử lớn hơn nhỏ nhất phía sau, sau đó đảo ngược đoạn đuôi để lấy hoán vị sát liền sau.

```python
def next_permutation(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    if i >= 0:
        j = n - 1
        while arr[j] <= arr[i]:
            j -= 1
        arr[i], arr[j] = arr[j], arr[i]
    left = i + 1
    right = n - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr
```

### W9A7: Kiểm tra vị trí trồng hoa hợp lệ
Quét qua luống đất và kiểm tra các điều kiện biên: ô hiện tại trống và cả hai ô bên cạnh (nếu có) cũng phải trống để tiến hành đặt hoa mới.

```python
def can_place_flowers(flowerbed, k):
    count = 0
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i - 1] == 0)
            right_empty = (i == n - 1) or (flowerbed[i + 1] == 0)
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
                if count >= k:
                    return True
    return count >= k
```

### W9A8: Lưu vết tọa độ di chuyển bằng từ điển băm
Theo dõi tọa độ (x, y) của Rô-bốt qua từng bước lệnh di chuyển R, L, U, D và kiểm tra sự trùng lặp ngay lập tức.

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0): True}
    for move in moves:
        if move == "R":
            x += 1
        elif move == "L":
            x -= 1
        elif move == "U":
            y += 1
        elif move == "D":
            y -= 1
        if (x, y) in visited:
            return True
        visited[(x, y)] = True
    return False
```

```python
def Median(nums):
    a, b, c, d, e = nums[0], nums[1], nums[2], nums[3], nums[4]
    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    
    if e < c:
        if b < c:
            if b > e:
                return b
            return e
        else:
            return c
    else:
        if b < c:
            return c
        else:
            if b < d:
                if b < e:
                    return b
                return e
            else:
                if d < e:
                    return d
                return e
```

### W9A10: Sắp xếp trộn kết hợp kỹ thuật hai con trỏ tìm khoảng cách gần nhất
Tự thiết lập hàm `merge_sort` để sắp xếp vị trí các ngôi nhà và lò sưởi, sau đó dùng hai con trỏ tìm bán kính tối thiểu phủ kín toàn bộ các hộ gia đình.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    while i < len(left):
        res.append(left[i])
        i += 1
    while j < len(right):
        res.append(right[j])
        j += 1
    return res

def findRadius(houses, heaters):
    sorted_houses = merge_sort(houses)
    sorted_heaters = merge_sort(heaters)
    max_radius = 0
    i = 0
    m = len(sorted_heaters)
    for house in sorted_houses:
        while i < m - 1 and abs(sorted_heaters[i + 1] - house) <= abs(sorted_heaters[i] - house):
            i += 1
        dist = abs(sorted_heaters[i] - house)
        if dist > max_radius:
            max_radius = dist
    return max_radius
```

### W9A13: Thuật toán tính bình phương logarit O(log exp)
Thực hiện chia nhỏ số mũ xuống một nửa ở mỗi bước đệ quy để tối ưu số lần nhân, đáp ứng yêu cầu không dùng vòng lặp hay toán tử mũ tích hợp sẵn.

```python
def power(base, exp):
    if exp == 0:
        return 1
    if exp == 1:
        return base
    half = power(base, exp // 2)
    if exp % 2 == 0:
        return half * half
    else:
        return half * half * base
```

### W9A14: Thống kê số lần xuất hiện của các phần tử

```python
def find_frequent_numbers(arr, x):
    counts = {}
    for num in arr:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
    res = []
    for num in counts:
        if counts[num] > x:
            res.append(num)
    if not res:
        return "INVALID"
    return res
```

### W9A15: Xác thực tính đơn điệu ngặt của dãy số
Duyệt qua mảng và kiểm tra đồng thời hai trạng thái tăng ngặt (phần tử sau luôn lớn hơn phần tử trước) và giảm ngặt (phần tử sau luôn nhỏ hơn phần tử trước).

```python
def check_strict_sequence(arr):
    if len(arr) <= 1:
        return "YES"
    is_inc = True
    is_dec = True
    for i in range(1, len(arr)):
        if arr[i] <= arr[i - 1]:
            is_inc = False
        if arr[i] >= arr[i - 1]:
            is_dec = False
    if is_inc or is_dec:
        return "YES"
    return "NO"
```
