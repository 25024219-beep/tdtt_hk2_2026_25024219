W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.

```python
pairs1 = input().split()
pairs2 = input().split()
d1 = {}
for p in pairs1:
    k, v = p.split(':')
    d1[k] = int(v)
d2 = {}
for p in pairs2:
    k, v = p.split(':')
    d2[k] = int(v)
res = {}
all_keys = set(d1.keys()).union(set(d2.keys()))
for k in sorted(all_keys):
    res[k] = d1.get(k, 0) + d2.get(k, 0)
print(res)
```

WW6A10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được

```python
nums = list(map(int, input().split()))
k = int(input())
pairs = []
n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))
pairs.sort(key=lambda x: x[0])
print(pairs)
```

W6A12
[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.

```python
nums = list(map(int, input().split()))
counts = {}
for x in nums:
    counts[x] = counts.get(x, 0) + 1
max_count = max(counts.values())
candidates = [k for k, v in counts.items() if v == max_count]
print(min(candidates))
```

W6A15 [Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới.

```python
pairs = input().split()
k = int(input())
d = {}
for p in pairs:
    key, val = p.split(':')
    d[key] = int(val)
res = {key: val for key, val in d.items() if val > k}
print(res)
```

## W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list


```python
def max_lst(lst):
    max_val = lst[0]
    max_index = 0
    for i in range(len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            max_index = i
    return max_index
lst = list(map(int, input().split()))
print(max_lst(lst))
```

## W6A20 44 
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
lst = list(map(int, input().split()))
k = int(input())
def tim_vi_tri(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
print(tim_vi_tri(lst, k))
```
