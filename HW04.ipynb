Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu kí tự số

```python
def dem_ki_tu(chuoi):
    hoa = 0
    thuong = 0
    # nếu là chữ hoa thì + 1 vào hoa, tương tự chữ thường +1 vaò thuong
    for chu in chuoi:
        if chu.isupper():
            hoa += 1
        elif chu.islower():
            thuong += 1
        
    return hoa, thuong

# nhập liệu
dau_vao = input()
hoa, thuong = dem_ki_tu(dau_vao)
print(f" có {hoa} kí tự in hoa")
print(f" có {thuong} kí tự in thường")
```

Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó
VD:
Nhập: 3, 12, 15
Tổng: 30 

```python
#nhập vào một dãy số nguyên, mỗi số cách nhau bởi dấu phẩy, sau đó tách dãy đó thành từng số, chuyển thành dạng int
so = [int(x) for x in input().split(",")]
tong = 0:
#khoá điều kiện đề bài
if len(so) == 3:
    for x in so:
        tong += x
    print(f"tổng 3 số trong chuỗi là {tong}")
else:
    print("nhập liệu không đúng yêu cầu")
```
