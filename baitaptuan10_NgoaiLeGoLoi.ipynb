W10A3: [ValueError - HappyNumbers] Nhập vào một số nguyên dương a, Kiểm tra xem a có phải là số hạnh phúc không. (Dùng try... except bắt lỗi ValueError thêm set)

```python
try:
    val = input()
    a = int(val)
    if a <= 0:
        print("False")
    else:
        seen = set()
        is_happy = True
        while a != 1:
            if a in seen:
                is_happy = False
                break
            seen.add(a)
            a = sum(int(c)**2 for c in str(a))
        print(is_happy)
except ValueError:
    print("Loi nhap lieu")
```

W10A5: Viết chương trình chuyển tất cả phần tử của một list cho trước về int. Nếu phần tử nào trong list không thể chuyển được thì in cảnh báo và bỏ qua phần tử đó

```python
def convert_list_to_int(my_list):
    result = []
    for item in my_list:
        try:
            result.append(int(item))
        except ValueError:
            print(f"Canh bao: Khong the chuyen doi {item}")
    return result
```

W10A10: Viết chương trình nhập vào phần tử N (N>0) và một mảng gồm N số nguyên dương. Một mảng là hợp lệ nếu nó gồm các số khác nhau. In ra tổng các phần tử nếu mảng đó là hợp lệ. Nếu mảng không hợp lệ, in ra "Mang khong hop le".

```python
try:
    n = int(input())
    arr = [int(x) for x in input().split()]
    if len(arr) == n and len(set(arr)) == len(arr) and all(x > 0 for x in arr):
        print(sum(arr))
    else:
        print("Mang khong hop le")
except ValueError:
    print("Mang khong hop le")
```

W10A14: [TypeError - SumNumbers] Nhập vào một danh sách gồm nhiều giá trị. Hãy cộng tất cả phần tử sau khi chuyển được sang int. Nếu phần tử nào không chuyển được thì bỏ qua. In ra tổng cuối cùng.

```python
def sum_valid_numbers(my_list):
    total = 0
    for item in my_list:
        try:
            total += int(item)
        except (ValueError, TypeError):
            pass
    return total
```

W10A15: [Raise_Exception - Positive Number] Nhập vào một số nguyên n. Nếu n<=0 thì chủ động phát sinh ngoại lệ và in ra "So phai duong". Nếu hợp lệ thì in ra bình phương của n.

```python
try:
    n = int(input())
    if n <= 0:
        raise Exception("So phai duong")
    print(n * n)
except Exception as e:
    if str(e) == "So phai duong":
        print("So phai duong")
    else:
        print("Loi luong du lieu")
```
