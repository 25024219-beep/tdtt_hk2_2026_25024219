
Bài 3: Nhập vào số nguyên dương a, in ra bảng cửu chương của a

```python
def bang_cuu_chuong(n):
    print(f"Bảng cửu chương của {n}:")
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")
#nhập số nguyên dương n
n = int(input("Nhập một số nguyên dương: "))
#kiểm tra nếu n là số nguyên dương
if n > 0:
    bang_cuu_chuong(n)
else:
    print("Vui lòng nhập một số nguyên dương.") 
```

Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python

#nhập chiều cao của tam giác cân
h = int(input("nhập chiều cao của tam giác cân"))
if h < 2:
    print("không đủ độ cao")
else:
#vẽ tam giác cân bằng dấu
    for i in range (1, h + 1):
        print(" " * (h - i), end="")
        if i == 1:
            print("*")
        elif i == h:
            print("*" * (2 * i - 1))
        else:
            khoang_trang_giua = " " * ( 2 * i - 3 )
            print("*" + khoang_trang_giua + "*")
            
```

Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python
# tạo hàm tìm ước dương
def uoc_duong_chung(h):
    uoc_duong = []
    for i in range(1, h + 1):
        if h % i == 0:
            uoc_duong.append(i)
#không thuộc điều kiện bỏ qua
        else:
            pass
    return  uoc_duong
#nhập liệu
a, b = map(int, input().split())
#dùng hàm đã code tìm danh sách ước dương của từng số
ds_a = uoc_duong_chung(a)
ds_b = uoc_duong_chung(b)
#so sánh 2 danh sách, trùng thì thêm vào ds_so_sanh
ds_so_sanh = []
for so in ds_a:
    if so in ds_b:
        ds_so_sanh.append(so)
#in ra kết quả
print(f" các ước của {a} là {ds_a}")
print(f" các ước của {b} là {ds_b}")
print(" -> danh sách ước chung là:", ds_so_sanh)
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
import math

a = int(input())
if a <= 0:
    print("số không hợp lệ")
else:
    b = math.isqrt(a)
    if b * b == a:
        print(f"{a} là số chính phương")
    else:
        print(f"{a} không là số chính phương")


```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())
if A <= 0:
    print("số không hợp lệ")
else:
    S = 0.0
    n = 0
    
    while S <= A:
        n += 1
        S += 1 / n
print(f"số n nhỏ nhất thoả mãn là: {n}")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
def fibonaci(h):
    f1 = 1
    f2 = 1
    while (f1 + f2) <= h:
        f1, f2 = f2, f1 + f2
    return f2
A = int(input())
if A <= 0:
    print("số không hợp lệ")
else:
    b = fibonaci(A)
    print(f"số fibonacci lớn nhất nhưng không vượt qua {A} là: {b}")
```
