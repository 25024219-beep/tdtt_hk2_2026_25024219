W9A2: Tìm số lớn thứ k trong ma trận tăng dần theo dòng và cột

```python
def find_kth_largest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    low = matrix[0][0]
    high = matrix[n-1][m-1]
    
    def count_greater_equal(mid):
        count = 0
        row = 0
        col = m - 1
        # Đếm số lượng phần tử >= mid với độ phức tạp O(n + m)
        while row < n and col >= 0:
            if matrix[row][col] >= mid:
                count += (col + 1)
                row += 1
            else:
                col -= 1
        return count

    ans = low
    while low <= high:
        mid = (low + high) // 2
        if count_greater_equal(mid) >= k:
            ans = mid
            low = mid + 1
        else:
            high = mid - 1
    return ans
```

W9A3: Hoán đổi đúng 2 chữ số để được số nhỏ nhất

```python
def min_swap_number(num_str):
    s = list(num_str)
    n = len(s)
    for i in range(n):
        for j in range(i + 1, n):
            if s[j] < s[i]:
                s[i], s[j] = s[j], s[i]
                # Chuyển về int để xóa số 0 ở đầu rồi chuyển lại thành string
                return str(int("".join(s)))
    return "".join(s)
```

W9A4: Biến đổi xâu ký tự

```python
def can_transform(a, b):
    n, m = len(a), len(b)
    dp = [[False] * (m + 1) for _ in range(n + 1)]
    dp[0][0] = True
    
    for i in range(1, n + 1):
        if a[i-1].islower():
            dp[i][0] = dp[i-1][0]
            
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if a[i-1].islower():
                # Có thể xóa chữ thường hoặc giữ lại nếu khớp sau khi viết hoa
                dp[i][j] = dp[i-1][j]
                if a[i-1].upper() == b[j-1]:
                    dp[i][j] = dp[i][j] or dp[i-1][j-1]
            else:
                # Chữ hoa bắt buộc phải khớp
                if a[i-1] == b[j-1]:
                    dp[i][j] = dp[i-1][j-1]
                    
    return "YES" if dp[n][m] else "NO"
```

W9A5: Tìm bán kính lò sưởi nhỏ nhất (Heaters)

```python
import bisect

def findRadius(houses, heaters):
    # Bắt buộc phải sắp xếp danh sách vị trí lò sưởi để tìm kiếm nhị phân
    heaters.sort()
    max_radius = 0
    
    for house in houses:
        # Tìm vị trí thích hợp của ngôi nhà trong danh sách lò sưởi đã sắp xếp
        idx = bisect.bisect_left(heaters, house)
        
        dist1 = float('inf')
        dist2 = float('inf')
        
        # Khoảng cách tới lò sưởi bên phải (hoặc ngay tại vị trí nhà)
        if idx < len(heaters):
            dist1 = heaters[idx] - house
            
        # Khoảng cách tới lò sưởi bên trái
        if idx > 0:
            dist2 = house - heaters[idx-1]
            
        # Khoảng cách tới lò sưởi gần nhất của ngôi nhà này
        min_dist = min(dist1, dist2)
        
        # Bán kính lò sưởi chung phải phủ được ngôi nhà có khoảng cách xa lò sưởi nhất
        max_radius = max(max_radius, min_dist)
        
    return max_radius
```

```python
def next_permutation(n, nums):
    i = n - 2
    # 1. Tìm vị trí i lớn nhất sao cho nums[i] < nums[i+1]
    while i >= 0 and nums[i] >= nums[i+1]:
        i -= 1
        
    if i >= 0:
        # 2. Tìm vị trí j lớn nhất sao cho nums[j] > nums[i]
        j = n - 1
        while nums[j] <= nums[i]:
            j -= 1
        # Hoán đổi hai phần tử
        nums[i], nums[j] = nums[j], nums[i]
        
    # 3. Đảo ngược đoạn phía sau i để thu được hoán vị nhỏ nhất kế tiếp
    nums[i+1:] = reversed(nums[i+1:])
    return nums
```

```python
def canPlaceFlowers(flowerbed, k):
    count = 0
    n = len(flowerbed)
    
    for i in range(n):
        if flowerbed[i] == 0:
            # Kiểm tra ô bên trái và ô bên phải xem có trống không
            left_empty = (i == 0) or (flowerbed[i-1] == 0)
            right_empty = (i == n - 1) or (flowerbed[i+1] == 0)
            
            if left_empty and right_empty:
                flowerbed[i] = 1  # Trồng hoa vào ô này
                count += 1
                if count >= k:
                    return True
                    
    return count >= k
```

W9A8: Phát hiện đường đi lặp lại của Rô-bốt

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}  # Tập hợp lưu tọa độ các ô đã đi qua
    
    for move in moves:
        if move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
            
        if (x, y) in visited:
            return True
        visited.add((x, y))
        
    return False
```

W9A9: Tìm trung vị của 5 số (Tối đa 6 phép so sánh, không dùng sort)

```python
def Median(nums):
    a, b, c, d, e = nums[0], nums[1], nums[2], nums[3], nums[4]
    
    # 3 phép so sánh nền tảng để thiết lập thứ tự cục bộ
    if a > b: a, b = b, a       # (1)
    if c > d: c, d = d, c       # (2)
    if a > c:
        a, c = c, a
        b, d = d, b             # (3) -> Đảm bảo a là min, và c < d
        
    # Nhánh quyết định tối ưu cây nhị phân để luôn đạt đúng 6 phép so sánh
    if e < c:                   # (4)
        if b < c:               # (5)
            return b if b > e else e  # (6)
        else:
            return c if b < d else d  # (6)
    else:                       # e >= c
        if b > d:               # (5)
            return e if e < d else d  # (6)
        else:                   # b <= d
            return e if e < b else b  # (6)
```

W9A10: Tìm bán kính lò sưởi nhỏ nhất (Heaters)

```python
import bisect

def findRadius(houses, heaters):
    heaters.sort()  # Sắp xếp danh sách lò sưởi để tìm kiếm nhị phân
    max_radius = 0
    
    for house in houses:
        # Tìm vị trí thích hợp của ngôi nhà trong mảng lò sưởi bằng bisect
        idx = bisect.bisect_left(heaters, house)
        
        # Khoảng cách tới lò sưởi bên phải
        dist1 = heaters[idx] - house if idx < len(heaters) else float('inf')
        # Khoảng cách tới lò sưởi bên trái
        dist2 = house - heaters[idx-1] if idx > 0 else float('inf')
        
        # Bán kính nhỏ nhất để phủ ngôi nhà này là khoảng cách tới lò sưởi gần nó nhất
        current_min_dist = min(dist1, dist2)
        
        # Bán kính tổng thể là giá trị lớn nhất trong các khoảng cách nhỏ nhất đó
        max_radius = max(max_radius, current_min_dist)
        
    return max_radius
```
